```
PATH  = Path('/content/drive/My Drive/Batik-StyleGAN/')
batik_raw_images = PATH / 'batik_raw_images/'
batik_preprocessed_images = PATH / 'batik_preprocessed_images'
```

```
batik_images = get_image_files(batik_raw_images)
len(batik_images)
```

```
for i in range(len(batik_images)):
  outfile = os.path.splitext(os.path.basename(batik_images[i]))[0]
  extension = os.path.splitext(batik_images[i])[1]
  im = PIL.Image.open(batik_images[i])
  rgb = im.convert('RGB')
  print(i+1," ",batik_images[i].stem," .......... done!")
  img_resized = rgb.resize((128,128))
  img_resized.save(os.path.join(batik_preprocessed_images, outfile+extension),"JPEG")
print("beres bossque")
```
